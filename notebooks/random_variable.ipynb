```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import axes3d, Axes3D #<-- Note the capitalization! 
```

# Discrete random variables

```python
state_set = ['AMZN', 'GOOG', 'FB']
p = [0.2, 0.5, 0.3]

sample_size = 10000
samples = np.random.choice(state_set, size=sample_size, p=p)

ratios = []
for i in range(len(state_set)):
    count = list(samples).count(state_set[i])
    ratios.append(list(samples).count(state_set[i]) / sample_size)

plt.bar(x=state_set, height=ratios)
```

# Continuous random variables

```python
fig = plt.figure()

ax = Axes3D(fig) 

# Make data.
S, A = np.mgrid[0:3:.01, 0:3:.01]
pos = np.empty(S.shape + (2,))
pos[:, :, 0] = S; pos[:, :, 1] = A
rv = multivariate_normal(mean=[1, 1], cov=[[3.0, 0.3], [0.3, 0.5]]) # rv means random variable
R = rv.pdf(pos)


# Plot the surface.
surf = ax.plot_surface(S, A, R, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_xlabel('States')
ax.set_ylabel('Actions')
ax.set_zlabel('Reward')
```
